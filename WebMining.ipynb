```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

url='http://127.0.0.1:3000/index.html'

r=requests.get(url)

soup=BeautifulSoup(r.text,'html.parser')
print(soup.prettify)
```

```python
dict={'Name':[],'Price':[],'Specification':[]}
```

```python
names=soup.findAll(class_='name')

for name in names:
    print(name.get_text())
    dict['Name'].append(name.get_text())

prices=soup.findAll(class_='price')

for price in prices:
    print(price.get_text())
    dict['Price'].append(price.get_text())

specifications=soup.findAll(class_='specification')

for specification in specifications:
    print(specification.get_text())
    dict['Specification'].append(specification.get_text().strip())
```

```python
df=pd.DataFrame(dict)
```

```python
print(df)
```

```python
df.to_excel("web.xlsx",index=False)
```
